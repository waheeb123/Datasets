FEMA Public Assistance Program
The Federal Emergency Management Agency (FEMA) Public Assistance (PA) Program is one of the primary federal disaster recovery programs in the United States. It provides financial assistance to state governments, local governments, tribal governments, and eligible nonprofit organizations following presidentially declared disasters. The program supports emergency response activities, debris removal, emergency protective measures, and the repair, restoration, or replacement of damaged public infrastructure. By reducing the financial burden on local communities, the PA Program helps accelerate disaster recovery and improve community resilience.
Disaster Recovery Funding
Effective disaster recovery depends heavily on the timely and equitable distribution of financial resources. Previous studies have shown that federal disaster assistance plays a significant role in restoring infrastructure, maintaining essential public services, and supporting long-term economic recovery. However, researchers have also found that funding allocation often varies according to disaster severity, geographic location, applicant characteristics, and administrative processes. These variations have motivated ongoing research into improving the transparency and efficiency of disaster funding.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('https://github.com/waheeb123/Datasets/blob/main/Sandy_PA_Master_Data_20260721.csv?raw=true')
```

```python
df.head()
```

```python
# Column names
df.columns
```

```python
# Data types
df.info()
```

```python
# Summary statistics
df.describe(include='all')
```

```python
# Remove State Share because it contains no data
df.drop(columns=["State Share"], inplace=True)
```

```python
df.info()
```

```python
money_columns = [
    "Obligated Amount",
    "Federal Share",
    "Local Share",
    "Federal Disbursed Amount"
]

for col in money_columns:
    # Convert to string first (handles missing values safely)
    df[col] = df[col].astype(str)

    # Remove dollar signs and commas
    df[col] = (
        df[col]
        .str.replace("$", "", regex=False)
        .str.replace(",", "", regex=False)
    )

    # Convert to numeric
    df[col] = pd.to_numeric(df[col], errors="coerce")
```
